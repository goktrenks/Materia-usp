```python
def name_conversion(name):
    if name == 'TOP':
        return 'top'
    if name == 'JUNGLE':
        return 'jg'
    if name == 'MIDDLE':
        return 'mid'
    if name == 'BOTTOM':
        return 'adc'
    if name == 'UTILITY':
        return 'sup'
```

```python
def get_position__and_team_from_column(column_name):
    split = column_name.split(sep = "_")
    return split[-2] + '_' + split[-1]
    
```

```python
def position_to_list_index(position, team):
    if team == 'blue':
        if position == 'top':
            return 0
        if position == 'jg':
            return 1
        if position == 'mid':
            return 2
        if position == 'adc':
            return 3
        if position == 'sup':
            return 4
    
    if team == 'red':
        if position == 'top':
            return 5
        if position == 'jg':
            return 6
        if position == 'mid':
            return 7
        if position == 'adc':
            return 8
        if position == 'sup':
            return 9
```

```python
i = 0
for line in (df.index):
    for feature in features:
        stat = df[feature][line]
        #name = name_conversion(get_position_from_column(feature))
        place_to_go = name_conversion(df[f"g_truth_{get_position__and_team_from_column(feature)}"][line])
        team_to_go = feature.split(sep = "_")[-1]
        column_to_go = position_to_list_index(place_to_go, team_to_go)
        new_df[i][column_to_go] = stat
        print(i)
    i += 1
```
