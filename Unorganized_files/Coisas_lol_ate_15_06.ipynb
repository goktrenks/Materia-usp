```python
def creating_id_and_tags(champ_names_list, data_dict, champ_and_ids_dict):
    data = data_dict
    champ_names = champ_names_list
    champ_and_ids = champ_and_ids_dict
    ids_and_tags = dict()
    
    for champ in champ_names:
        try:
            id_ = champ_and_ids[champ]
            ids_and_tags[id_] = data['data'][champ]['tags']
        except KeyError:
            continue
            
    return ids_and_tags
```

```python
id_and_tags = creating_id_and_tags(champ_names, data, champ_and_ids)
```
